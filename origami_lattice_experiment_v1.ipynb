## 3. Baseline Lattice Reduction

```python
def vector_norms(B):
    return [np.linalg.norm(v) for v in B]

def shortest_vector_length(B):
    return min(vector_norms(B))
```

## 4. Origami Fold Operators

```python
import numpy as np

def reflect_vector(v, normal):
    normal = np.array(normal, dtype=float)
    v = np.array(v, dtype=float)
    normal = normal / np.linalg.norm(normal)
    return v - 2 * np.dot(v, normal) * normal

def fold_basis(B, normal):
    return np.array([reflect_vector(v, normal) for v in B])
```

## 6. Metrics

```python
import numpy as np
import math

def shortest_combination_length(B, coeff_range=3):
    best_len = float("inf")
    best_coeffs = None
    best_vec = None

    for c1 in range(-coeff_range, coeff_range + 1):
        for c2 in range(-coeff_range, coeff_range + 1):
            if c1 == 0 and c2 == 0:
                continue

            vec = c1 * B[0] + c2 * B[1]
            length = np.linalg.norm(vec)

            if length < best_len:
                best_len = length
                best_coeffs = (c1, c2)
                best_vec = vec

    return best_len, best_coeffs, best_vec
```

```python
import numpy as np
import random
import statistics

def random_lattice():
    a = random.randint(5, 50)
    b = random.randint(1, 20)
    return np.array([[a, 0.],
                     [b, 1.]])

def basis_difference_normal(B):
    v1 = B[0]
    v2 = B[1]
    normal = v1 - v2
    if np.linalg.norm(normal) == 0:
        normal = np.array([1.0, 0.0])
    return normal

def run_smart_experiment(n_trials=100):
    improvements = 0
    results = []

    for _ in range(n_trials):
        B = random_lattice()

        normal = basis_difference_normal(B)
        B_folded = fold_basis(B, normal)

        original_short = shortest_vector_length(B)
        folded_short = shortest_vector_length(B_folded)

        improved = folded_short < original_short

        if improved:
            improvements += 1

        results.append({
            "original": original_short,
            "folded": folded_short,
            "improved": improved
        })

    return improvements, results

improvements, results = run_smart_experiment(100)

print("Total improvements:", improvements)
print("Improvement rate:", improvements / 100)

originals = [r["original"] for r in results]
foldeds = [r["folded"] for r in results]

print("Average original:", statistics.mean(originals))
print("Average folded:", statistics.mean(foldeds))
```
